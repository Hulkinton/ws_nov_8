```python
masses = [3.5, 2.1, 9.2, 1.7]

for mass in masses:
    if mass > 9.0:
        print(mass, "is HUGE")
        
    elif mass > 3.0:
        print(mass, "is large")
        
    else:
        print(mass, "is small")
print(mass)
```

```python
original = [-1.5, 0.2, 0.4, 0.0, -1.3, 0.4]

result = []

for value in original:
    if value < 0:
        result.append(0)
    else:
        result.append(1)
        
print(result)
```
